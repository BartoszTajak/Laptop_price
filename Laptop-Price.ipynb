```python
# Load data from Postgresql
db = create_engine("postgresql://postgres:komandos08@localhost:5432/sound")
conn = db.connect()
df_sql = pd.read_sql('SELECT * FROM laptop_price', conn  )
```

```python
# quick view how dataframe looks before converting
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df_sql['Price_euros']   , color = 'blue', kde=True , bins =40,norm_hist = True)
```

```python
#Delete this column because there is no impact to results
df_sql.drop(['Product'], axis=1, inplace=True)
```

```python
# create new value based on resolution from 'ScreenResolution'
New_FeturesY= df_sql['ScreenResolution'].apply(lambda data: int(data.split('x')[-1:][0]))
New_FeturesX = df_sql['ScreenResolution'].apply(lambda data: int(data.split('x')[0][-4:]))

# change columns based on New_FeturesY and New_FeturesX (float value instead string and one hot...)
df_sql['ScreenResolution'] = (New_FeturesY * New_FeturesX)/1000
```

```python
# remowing rows with value_counts < 10
for i in ['TypeName','Gpu','OpSys','TypeName']:  
    df_sql_series = df_sql[i].value_counts()<10
    lista = df_sql_series.index[df_sql_series].to_list()
    df_sql = df_sql[~df_sql[i].isin(lista)]
```

```python
# removing rows consists misfit values
df_sql = df_sql[(df_sql['Price_euros'] <2900)]
```

```python
# quick view how dataframe looks after converting (now is more symetric)
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df_sql['Price_euros']   , color = 'blue', kde=True , bins =40,norm_hist = True)
```

```python
df_sql['Weight'].str.contains('kg').sum()
```

```python
df_sql['Ram'].str.contains('GB').sum()
```

```python
df_sql = df_sql.astype({"Ram": str, "Weight": str})
# remove GB and kg from columns
df_sql['Ram'] = df_sql['Ram'].apply(lambda data: data.replace('GB', ''))
df_sql['Weight'] = df_sql['Weight'].apply(lambda data: data.replace('kg', ''))
# convert column "Ram": int, "Weight": float
df_sql = df_sql.astype({"Ram": int, "Weight": float})
```

```python
df_sql['Cpu_model'] = df_sql['Cpu'].str.extract(r'(i\d|AMD|Samsung|Pentium|Celeron|Atom|Core M|Xeon)')
df_sql['Cpu_freq'] = df_sql['Cpu'].str.extract(r'(\d+\.\d+GHz|\d+GHz)')
df_sql['Cpu_freq'] = df_sql['Cpu_freq'].str.replace('GHz','').astype('float64')
df_sql['Memory_type'] = df_sql['Memory'].str.extract('(SSD|Flash |HDD)')
df_sql['Memory_Size'] = df_sql.Memory.str.extract(r'(\d+GB|\d+TB)')
df_sql['Memory_Size'] = df_sql.Memory_Size.str.extract(r'(\d+)').astype('int64')

df_sql.drop(['Cpu'], axis=1, inplace=True)
df_sql.drop(['Memory'], axis=1, inplace=True)
```

```python
# One hot encode and add to frame
df_sql = pd.get_dummies(df_sql, columns=['Company', 'TypeName','Gpu','OpSys','Cpu_model','Memory_type'])
```

```python
# reorder index
df_sql.index = np.arange(1, len(df_sql) + 1)
```

```python
# Try to create new features
df_sql['New_Fetures_two'] = df_sql['ScreenResolution']/df_sql['Inches']/df_sql['Weight']
```

```python
# dropping duplicate columns and create X and Y frame
X = df_sql.drop(columns=['Price_euros'])
Y = df_sql['Price_euros']
Y = Y.to_frame()
```

```python
# few lines to remove features with high correlation between each other
corr_matrix = X.corr().abs()
upper_tri_corr = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
triu_cols = upper_tri_corr.columns
high_correlated_features = [column for column in triu_cols if any(upper_tri_corr[column] > 0.9)]
X = X.drop(columns=high_correlated_features)
```

```python
# split data Train and Test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=68)
min_max_scaler = MinMaxScaler()
X_train_norm = min_max_scaler.fit_transform(X_train)
X_test_norm = min_max_scaler.transform(X_test)


# PCA method has a low efficient
# pca = PCA(n_components=300)
# # Transformacja
# x_train_norm_PCA = pca.fit_transform(X_train_norm)
# x_test_norm_PCA = pca.transform(X_test_norm)
```

```python
gboost_reg = xgb.XGBRegressor()
gboost_reg.fit(X_train_norm, Y_train)
Y_predictions_test = gboost_reg.predict(X_test_norm)
MSE_test_gboost_reg =round(metrics.mean_squared_error(Y_test, Y_predictions_test, squared=False), 2)
print('Metrics mean squared error = ', MSE_test_gboost_reg)
plt.figure(figsize=(12, 8), dpi=80)
plt.xlabel("Number of sample")
plt.ylabel("Price in euro")
plt.scatter(list(range(len(Y_predictions_test))), Y_predictions_test, c='y')
plt.scatter(list(range(len(Y_test))), Y_test, c='r')
```

```python
RandomForest = RandomForestRegressor(n_estimators = 30)
RandomForest.fit(X_train_norm, Y_train.values.ravel())
Y_predictions_train = RandomForest.predict(X_train_norm)
Y_predictions_test = RandomForest.predict(X_test_norm)
MSE_test_RandomForest =round(metrics.mean_squared_error(Y_test, Y_predictions_test, squared=False), 2)
print('Metrics mean squared error = ', MSE_test_RandomForest)
plt.figure(figsize=(12, 8), dpi=80)
plt.xlabel("Number of sample")
plt.ylabel("Price in euro")
plt.scatter(list(range(len(Y_predictions_test))), Y_predictions_test, c='y')
plt.scatter(list(range(len(Y_test))), Y_test, c='r')
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(300, activation='relu', input_shape=(X_train_norm.shape[1],)))
model.add(tf.keras.layers.Dense(200, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer='adam', loss='mean_absolute_percentage_error')
model.fit(X_train_norm, Y_train, epochs=200, verbose=1)
Y_predictions_test = model.predict(X_test_norm)
MSE_test_Neural_Network = round(metrics.mean_squared_error(Y_test, Y_predictions_test, squared=False),2)
print('Metrics mean squared error = ', MSE_test_Neural_Network)
plt.figure(figsize=(12, 8), dpi=80)
plt.xlabel("Number of sample")
plt.ylabel("Price in euro")
plt.scatter(list(range(len(Y_predictions_test))), Y_predictions_test, c='y')
plt.scatter(list(range(len(Y_test))), Y_test, c='r')

```
